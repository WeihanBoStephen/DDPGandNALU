# This code is for visualisation of DDPG loss function

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
actor_loss = pd.read_csv('./Loss_Function_File/actor_loss.csv')
actor_loss
```

```python
losses = { 'epoch': []}
```

```python
losses['epoch'] = actor_loss['Value'].values
```

```python
loss_type = 'epoch'
iters = actor_loss['Step'].values
plt.figure(figsize=(8,8))
plt.title("Actor Training Loss epoch-0~epoch5759",fontsize=18)
plt.plot(iters, losses[loss_type], 'g', label='train loss')

plt.grid(True)#è®¾ç½®ç½‘æ ¼å½¢å¼
plt.xlabel(loss_type,fontsize=16)
plt.ylabel('loss',fontsize=16)#ç»™xï¼Œyè½´åŠ æ³¨é‡Š
plt.legend(loc="upper right")#è®¾ç½®å›¾ä¾‹æ˜¾ç¤ºä½ç½®
plt.savefig("./Loss_Function_File/actor_loss_0_5759.jpg")
plt.show()
```

```python
ddpg_test = pd.read_csv('./Model_Test_Results/ddpg_model_test01.csv')
ddpg_test
```

```python
nalu_test = pd.read_csv('./Model_Test_Results/ddpg_nalu_test01.csv')
nalu_test
```

```python
testes = { 'step': [],'nalu':[]}
testes['nalu'] = nalu_test['step'].values
testes['step'] = ddpg_test['step'].values
loss_type = 'step'
iters = range(len(testes[loss_type]))
plt.figure(figsize=(8,8))
plt.title("DDPG vs DDPG&NALU",fontsize=18)
plt.plot(iters, testes[loss_type], 'g', label='Normal DDPG')
plt.plot(iters, testes['nalu'], 'r', label='DDPG with NALU')
plt.grid(True)
plt.xlabel("Wechat Game",fontsize=16)
plt.ylabel('step',fontsize=16)#ç»™xï¼Œyè½´åŠ æ³¨é‡Š
plt.legend(loc="upper right")#è®¾ç½®å›¾ä¾‹æ˜¾ç¤ºä½ç½®
plt.savefig("./Model_Test_Results/ddpg_model_test01.jpg")
plt.show()
```
