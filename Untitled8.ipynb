# This is NALU Training code

```python
import numpy as np
import keras.backend as K
from keras.layers import *
from keras.initializers import *
from keras.models import *
import pandas as pd
import keras
import matplotlib.pyplot as plt
from keras.callbacks import ModelCheckpoint
import nalu_model as nalu
import pandas as pd
from sklearn.cross_validation import train_test_split
```

```python
def presstime_to_action(presstime):
    action = (presstime - 200) / 600 - 1
    return action
```

```python
data = pd.read_csv('nalu_data.csv')
print(len(data))
data
```

```python
train, test = train_test_split(data, test_size = 0.2)
trX = train.drop(['Unnamed: 0','ddpg_presstime','opencv_presstime','img'],axis=1).values
trY = train['opencv_presstime'].values
teX = test.drop(['Unnamed: 0','ddpg_presstime','opencv_presstime','img'],axis=1).values
teY = test['opencv_presstime'].values
trY = presstime_to_action(trY)
teY = presstime_to_action(teY)
```

```python
print(trX.shape)
print(trY.shape)
print(teX.shape)
print(teY.shape)
trY
```

```python
teY
```

```python
filepath="./nalu_experiments/model_weights01.h5"
```

```python
m = nalu.get_model("NALU")
history = nalu.get_history()
m.compile("rmsprop", "mse", metrics=["mae"])
K.set_value(m.optimizer.lr, 1e-2)
m.fit(trX, trY, validation_data=(teX, teY), batch_size=100, epochs=1000,callbacks=[history])
K.set_value(m.optimizer.lr, 1e-3)
m.fit(trX, trY, validation_data=(teX, teY), batch_size=100, epochs=1000,callbacks=[history])
K.set_value(m.optimizer.lr, 1e-4)
m.fit(trX, trY, validation_data=(teX, teY), batch_size=100, epochs=1000,callbacks=[history])
m.save_weights(filepath)
```

```python
loss,accuracy = m.evaluate(teX,teY)
print('\ntest loss',loss)
print('accuracy',accuracy)
```
